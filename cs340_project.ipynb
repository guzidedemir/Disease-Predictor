```
#symptom_list
from pyspark.sql import Row
from pyspark.ml.linalg import Vectors
import random
predicted_value_list=[]
for i in range(100):
  
  for i in range(20):
    combination=[]
    b_combination=[]
    random2=random.choice(symptom_indexes)
  
    combination.append(random2)
    for i in range(403):
      if i in combination:
        b_combination.append(1)
      else:
        
        b_combination.append(0)
    
  rdd_user= sc.parallelize(b_combination)

  df3 = sqlContext.createDataFrame([Row(  features = Vectors.dense([float(i)for i in b_combination ]))])

  prediction = model.transform(df3)
  a=prediction.select("prediction").collect()
  predicted_value=a[0][0]
  predicted_value_list.append(predicted_value)
print(predicted_value_list)
```

```
list_rdd= sc.parallelize(predicted_value_list)
list_length=len(predicted_value_list)
k=list_rdd.map(lambda x: (x,1))
rdd=k.reduceByKey(lambda x,y:(x+y))
probability=[]
indexes_d_name=[]
for i in rdd.collect():
  probability.append((i[0],i[1]/list_length))
  indexes_d_name.append(i[0])
print(probability)
#print(indexes_d_name)
```

```
matched_diseases_names=[]
for i in indexes_d_name:
  for j in liste_numara :
    if (float(j[0]) == i):
      matched_diseases_names.append((j[0],j[1]))
#print(matched_diseases_names)


last_list=[]
for i in matched_diseases_names:
  for j in probability:
    #print(i[1],j)
    if ( float(i[0])==float(j[0]) ):
      last_list.append((i[1],j[1]))
print(last_list)


```

```
rdd_last=sc.parallelize(last_list)
df=sqlContext.createDataFrame(rdd_last)
df_last=df.withColumnRenamed('_1','disease names').withColumnRenamed('_2','probablities').sort("probablities",ascending=False)
df_last.show()
```
